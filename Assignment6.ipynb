```python
import pandas as pd
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

Naive bayes is a classification model and is based on probability.

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
df.head()
```

```python
df = df[df['Class Name']== 'Dresses']
df['review'] = (df['Rating'] > 3).astype(str)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

text = df['Review Text'].values.astype('U') 
vect = CountVectorizer(stop_words='english') 
vect = vect.fit(text) 
feature_names = vect.get_feature_names()
print(f"There are {len(feature_names)} words in the vocabulary. A selection:{feature_names[500:520]}")
```

```python
docu_feat = vect.transform(text)
```

```python
print(docu_feat[0:500,0:500])
```

```python
df_words = pd.concat([df[0:50].reset_index(), pd.DataFrame(docu_feat.toarray()[0:50]).reset_index()], axis=1)
df_words.head(5)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB() 
X = docu_feat 
y = df['review'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data and store it

nb = nb.fit(X_train, y_train) #fit the model X=features, y=character

#Evaluate the model
y_test_p = nb.predict(X_test)
nb.score(X_test, y_test)
```

```python
df['review'].value_counts(normalize=True)
```

```python
cm = confusion_matrix(y_test, y_test_p)
cm = pd.DataFrame(cm)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_p))
```
